```c++
TCanvas *c1 = new TCanvas();

TFile *input = new TFile("data_A.exactly2lep.root", "read");
TTree *tree = (TTree*)input->Get("mini");
std::vector<float> *vecE = nullptr;
tree->SetBranchAddress("lep_E",&vecE);
int entries = tree->GetEntries();

TH1F *histE = new TH1F("hist", "Histogram", 100, 0, 1000);
float E;
```

```c++
//tree->GetEntry(0);
//E = vecE -> at(0);
//cout << E/1000 << endl;
```

```c++
for(int i = 0; i < entries; i++)
{
tree->GetEntry(i);
E = vecE -> at(0);
//cout << E/1000 << endl;
histE->Fill(E/1000);
}
cout << E/1000 << endl;
```

```c++
histE->Draw();
c1->Draw();
```
